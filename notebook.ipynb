# Data Acquisition

First, we will load the dataset

Import necessary libraries

```python
import pandas as pd
```

Load datasets and display the first few rows of the dataset

```python
train_data = pd.read_csv('train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('test.csv')
test_data.head()
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Info


```python
train_data.info()
```

Basic statistics of the dataset

```python
train_data.describe()
```

Plot the distribution of price range

```python
sns.countplot(x='price_range', data=train_data)
plt.title('Distribution of Price Range')
plt.show()
```

we see that 4 classes are balanced

Correlation heatmap

# Data Preprocessing


```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

Handle Missing Values

```python
train_data.isnull().sum()
```

Separate features and target variable

```python
X= train_data.drop('price_range', axis=1)
y = train_data['price_range']
```

Feature Scaling

Normalize the features to improve model performance.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Split the data into training and validation sets

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train.shape, X_val.shape
```

# Model Training

We'll train a RandomForestClassifier as it's suitable for classification problems and often performs well without extensive tuning.

```python
from sklearn.ensemble import RandomForestClassifier
```

Train the RandomForestClassifier

```python
model = RandomForestClassifier(n_estimators=50, random_state=42)
model.fit(X_train, y_train)
model
```

Predict on the validation set

```python
y_train_preds= model.predict(X_train)
y_val_preds = model.predict(X_val)

train_data['y_train_preds'] = pd.DataFrame(y_train_preds)
train_data['y_val_pred'] = pd.DataFrame(y_val_preds)
train_data.head()
```

# Model Evaluation

Evaluate the model using confusion matrix, classification report, and accuracy score.

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

Accuracy score

```python
train_accuracy = accuracy_score(y_train, y_train_preds)
val_accuracy = accuracy_score(y_val, y_val_preds)
print('Train accuracy:', train_accuracy)
print('Val accuracy:', val_accuracy)
```

Classification report-

```python
cr = classification_report(y_val, y_val_preds)
print('Classification Report:\n', cr)
```

Confusion matrix

```python
cm = confusion_matrix(y_val, y_val_preds)
print('Confusion Matrix:\n', cm)
```

Confusion Matrix Display

```python
cmd = ConfusionMatrixDisplay(cm, display_labels=model.classes_)
cmd.plot()
```

### We will improve model more

# Feature Selection

Let's first analyze feature importance and remove less important features.

```python
importances = model.feature_importances_
importances
```

```python
features = X.columns
features
```

Create a DataFrame of feature importances

```python
feature_importance_df = pd.DataFrame({'Feature': features, 
                                      'Importance': importances})
feature_importance_df.head()
```

Sort feature importances by importance value

```python
feature_importance_df = feature_importance_df.sort_values(by='Importance', 
                                                          ascending=False)
feature_importance_df
```

Plotting feature importances

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance')
plt.show()
```

Concatenate feature importances into train_data

```python
train_data = pd.concat([train_data, feature_importance_df], axis=1)
train_data.head()
```

# Optimizing the Model

To further improve the model, we can tune hyperparameters using GridSearchCV

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
```

Defining the pipeline with scaling, feature selection and model

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('feature_selection', SelectFromModel(RandomForestClassifier(random_state=42))),
    ('classifier', RandomForestClassifier(random_state=42))
])
pipeline
```

Perform cross-validation scores

```python
cv_scores = cross_val_score(pipeline, X, y, cv=5)
cv_scores
```

Mean cross-validation scores

```python
cv_scores.mean()
```

Defining the parameter grid for GridSearchCV

```python
param_grid = {
    'feature_selection__estimator__n_estimators': [50, 100, 200],
    'feature_selection__estimator__max_depth': [None, 10, 20],
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
    'classifier__max_features': ['auto', 'sqrt', 'log2']
}
param_grid
```

Performing GridSearchCV with cross-validation

```python
grid_search = GridSearchCV(estimator=pipeline, 
                           param_grid=param_grid, 
                           cv=5, 
                           n_jobs=-1, 
                           verbose=2)
grid_search.fit(X_train, y_train)
grid_search
```

Getting the best parameters and best estimator

```python
best_params = grid_search.best_params_
best_params
```

Training the optimized model with best parameters

```python
optimized_model = grid_search.best_estimator_
optimized_model.fit(X_train, y_train)
optimized_model
```

Predict on the validation set with optimized model

```python
y_optimized_pred = optimized_model.predict(X_val)
train_data['y_val_optimized_pred'] = pd.DataFrame(y_optimized_pred)
train_data.head()
```

Evaluate optimized model

```python
optimized_accuracy = accuracy_score(y_val, y_optimized_pred)
print('Optimized Accuracy:', optimized_accuracy)
```

```python
import joblib
joblib.dump(optimized_model, 'optimized_model.pkl')
```
